```python
#checking p-stats based on alpha = 0.05
for col in raw_data.columns[1:]:
    stat, p_value = stats.shapiro(raw_data[col])
    alpha = 0.05
    if p_value > alpha:
        result = colored('Accepted', 'green')
    else:
        result = colored('Rejected', 'red')    
    print(col,result, p_value)
```

```python
#checking shapiro-wilk test for normality for p-stats based on alpha = 0.05
for col in raw_data.columns[1:]:
    stat, p_value = stats.kstest(raw_data[col], 'norm')
    alpha = 0.05
    if p_value > alpha:
        result = colored('Accepted', 'green')
    else:
        result = colored('Rejected', 'red')    
    print(col,result, p_value)
```

## Finding Best K-Means Clustering Model

```python
from sklearn.preprocessing import RobustScaler
robustscaler = RobustScaler()
data_robust = robustscaler.fit_transform(raw_data)
```

```python
from sklearn.preprocessing import StandardScaler
standardscaler = StandardScaler()
data_standard = standardscaler.fit_transform(raw_data)
```

```python
from sklearn.preprocessing import PowerTransformer

power_data_robust = PowerTransformer().fit(data_robust)
power_data_robust_transformed = power_data_robust.transform(data_robust)

power_data_standard = PowerTransformer().fit(data_standard)
power_data_standard_transformed = power_data_standard.transform(data_standard)
```

```python
from sklearn.cluster import KMeans

sse_robust = []
sse_standard = []
sse_robust_transformed = []
sse_standard_transformed = []

k_rng = range(1,15)
seed = 1

print('K-Val', '\t', 'Robust Score', '\t\t', 'Stand Score', '\t\t', 'PTrans+Robust', '\t\t', 'PTrans+Stand')

for k in k_rng:
    
    km_robust = KMeans(n_clusters=k, max_iter=300, random_state=seed)
    km_standard  = KMeans(n_clusters=k, max_iter=300, random_state=seed)
    km_robust_transformed  = KMeans(n_clusters=k, max_iter=300, random_state=seed)
    km_standard_transformed = KMeans(n_clusters=k, max_iter=300, random_state=seed)
    
    km_robust.fit(data_robust)
    km_standard.fit(data_standard)
    km_robust_transformed.fit(power_data_robust_transformed)
    km_standard_transformed.fit(power_data_standard_transformed)
    
    print('k=',k, '\t', km_robust.inertia_, '\t', km_standard.inertia_,
          '\t', km_robust_transformed.inertia_, '\t', km_standard_transformed.inertia_)
    
    sse_robust.append(km_robust.inertia_)
    sse_standard.append(km_standard.inertia_)
    sse_robust_transformed.append(km_robust_transformed.inertia_)
    sse_standard_transformed.append(km_standard_transformed.inertia_)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
l1=ax.plot(k_rng,sse_robust,'ys-')
l2=ax.plot(k_rng,sse_standard,'r--')
l3=ax.plot(k_rng,sse_robust_transformed,'b')
l4=ax.plot(k_rng,sse_standard_transformed,'go--')
ax.set_title('Elbow Curve From Different K-Means Models')
ax.set_xlabel('K Value')
ax.set_ylabel('SSE')
```

#### from above model fits and interpretation of the inertia it looks like the robust scaled data produces better results in k=8. Hence below is the prediction of the model

```python
km_robust = KMeans(n_clusters=8, max_iter=300, random_state=seed)
km_robust.fit(data_robust)
prediction = km_robust.predict(data_robust)
prediction
```
